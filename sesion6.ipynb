```python
import scipy.stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('bicicletas-compartidas.csv')
```

VARIABLES CATEGORICAS

```python
df.columns
```

```python
y = df['cuartil-ano'].apply(lambda x: 'Cuartil - '+str(int(x)))
```

```python
#fig=figuras , ax=ejes
fig, ax = plt.subplots()
ax.bar(y.value_counts().index,y.value_counts())
ax.set_xlabel('Cuartiles del año')
ax.set_ylabel('Frecuencia')
```

```python
#fig=figuras , ax=ejes
fig, ax = plt.subplots()
ax.bar(y.value_counts().index,y.value_counts())
ax.set_xlabel('Cuartiles del año')
ax.set_ylabel('Frecuencia')
ax.patches[3].set_facecolor('red')#coloreo el ultimo cuartil
ax.patches[2].set_facecolor('blue')#coloreo el ultimo cuartil
```

```python
fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels=y.value_counts().index)
ax.set_title('Diagrama de pie')
```

VARIABLES NUMERICAS

```python
y = df['viento']
```

```python
fig, ax = plt.subplots()
ax.hist(y,bins=30)
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuencias')
ax.set_title('Viento vs Frecuencia')
plt.axvline(np.mean(y),c='r',linestyle = '--', label = 'promedio')
plt.axvline(np.mean(y)+np.std(y) ,c='k',linestyle = '--', label = '+1 dev estandar')
plt.axvline(np.mean(y)-np.std(y) ,c='k',linestyle = '--', label = '-1 dev estandar')
ax.legend()
```

```python
y = df['bicis-compartidas']
fig,ax = plt.subplots()
ax.boxplot(x = y)
```

```python
#fig,ax = plt.subplots()
sns.boxplot(x = 'cuartil-ano', y = 'bicis-compartidas', data = df)
```

```python
fig,ax = plt.subplots()
ax.scatter(df['viento'],df['bicis-compartidas'],alpha = 0.03)
ax.set_title('distribucion conjunta')
ax.set_xlabel('Viento')
ax.set_ylabel('Bicis-Compartidas')
```
