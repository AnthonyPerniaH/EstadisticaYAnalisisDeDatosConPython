```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('iris-data.csv', index_col=0)
```

```python
df.columns
```

```python
df.tipo_flor.value_counts()
```

```python
y = df['lar.petalo']
```

```python
fig, ax = plt.subplots()
ax.set_title('Variable original')
ax.hist(y, bins = 30)
ax.axvline(x = np.mean(y), c = 'r', label = 'media', linestyle = '--')
ax.axvline(x = np.mean(y)+np.std(y), c = 'y', label = 'desviacion est.', linestyle = '--')
ax.legend()
```

*ESTANDARIZANDO LA VARIABLE*



```python
#Centrando, se resta la media a cda uno de los valores originales

fig, ax = plt.subplots()
ax.set_title('Variable original')
ax.hist(y-np.mean(y), bins = 30)
ax.axvline(x = np.mean(y- np.mean(y)), c = 'r', label = 'media', linestyle = '--')
ax.axvline(x = np.mean(y)+np.std(y), c = 'y', label = 'desviacion est.', linestyle = '--')
ax.legend()
```

```python
#reducir la variable, dividir todos sus valores por la desviacion

fig, ax = plt.subplots()
ax.set_title('Variable Estandarizada')
ax.hist((y-np.mean(y))/np.std(y), bins = 30)
ax.axvline(x = np.mean(np.mean(y- np.mean(y))/np.std(y)), c = 'r', label = 'media', linestyle = '--')
ax.axvline(x = np.mean(np.mean(y- np.mean(y))/np.std(y)) + np.std((y - np.mean(y))/np.std(y)), c = 'y', label = 'desviacion est.', linestyle = '--')
ax.legend()
```
