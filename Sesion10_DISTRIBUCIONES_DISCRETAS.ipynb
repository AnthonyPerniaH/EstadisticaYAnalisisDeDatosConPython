```python

import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import seaborn as sns
%matplotlib inline
```

```python
p = 0.3 #probabilidad de exito
data = bernoulli.rvs(p, size = 100)
len(data)
```

```python
mean, var, skew , kurt = bernoulli.stats(p, moments = 'mvsk')
mean, var, skew , kurt
```

```python
ax = sns.distplot(data , bins = 30, kde = False, color = 'blue')
ax.set(xlabel='distribucion de bernoulli', ylabel = 'frecuencia')
ax.axvline(x = mean, linestyle = '--', label = 'Media',color='r')
ax.legend()
```
