```python
import scipy.stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom
%matplotlib inline
```

```python
p = 0.3
n =8
x = list(range(0,9,1)) # sera de 0 a 8 , porque en python el ultimo valor no lo incluye
y = list([])

for i in x:
    y.append( binom.pmf(i, p = p, n = n) )
```

```python
    
fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('Probabilidad dicreta')
```

```python
np.sum(y)
```

```python
media = np.average(x, weights = y)
```

```python
varianza = np.average( ((x-media)**2), weights = y)
```

```python
media, varianza
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('Probabilidad dicreta')
ax.axvline( x = media, c = 'r', linestyle = '--', label = 'Valor esperado')
ax.axvline( x = media + 3*np.sqrt(varianza), c = 'r', linestyle = '--', label = 'Desviacion estandar')
ax.legend()
```
